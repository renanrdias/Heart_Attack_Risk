```python
# renomeando as colunas
df.rename(columns={column: column.lower().replace(" ", "_") for column in df.columns}, inplace=True)
```

```python
# Tratamento da coluna blood_pressure

pressures = df['blood_pressure'].str.split("/", expand=True).astype("int64")

df.insert(5, 'systolic', pressures[0])
df.insert(6, 'diastolic', pressures[1])

df.insert(7, 'systolic_diastolic_ratio', df['systolic'] / df['diastolic'])

df.drop("blood_pressure", axis=1, inplace=True)
```

```python
df.info()
```

```python
# Tratamento da coluna sex: Considerando Female = 1 e Male = 0

df["sex"] = np.where(df["sex"] == "Female", 1, 0)
```

```python
df.info()
```

```python
# tratamento da coluna hemisphere

# print(df.hemisphere.unique())

df["hemisphere"] = np.where(df["hemisphere"] == "Northern Hemisphere", 1, 0)
```

```python

data = df['country'].value_counts()
data = data.sort_values(ascending=False)

plt.figure(figsize=(10, 6))
bars = sns.barplot(x=data.index, y=data.values, palette='viridis')

plt.title('Country Distribution', size=18)
plt.xlabel('Country', size=14)
plt.ylabel('Count', size=14)
plt.xticks(rotation=45, ha='right')

for bar in bars.patches:
    height = bar.get_height()
    plt.annotate(f'{int(height)}', (bar.get_x() + bar.get_width() / 2, height), ha='center', va='bottom')
plt.show()
```
